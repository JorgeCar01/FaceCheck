```python
import matplotlib.pyplot as plt

train_loss = [1.3649, 1.1953, 1.0354, 0.9134, 0.7313, 0.5910, 0.4600, 0.3289, 0.2420, 0.1606, 0.5510]

train_acc = [38.3333, 63.3333, 90.8333, 97.5000, 100.0000, 100.0000, 100.0000, 100.0000, 100.0000, 100.0000]
```

```python
x = range(1, len(train_acc) + 1)

plt.plot(x, train_loss, label='Train Loss')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss per Epoch')
plt.legend()
plt.savefig('loss.png')
plt.show
```

```python
plt.plot(x, train_acc, label='Train Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy per Epoch')
plt.legend()
plt.savefig('accuracy.png')
plt.show
```
